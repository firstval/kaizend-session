```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

```python
fn="../output/wikipedia_metrics.csv"
df=pd.read_csv(fn)
```

```python
df.head()
```

```python
x=df["users"]
y=df["edits"]
```

```python
plt.plot(x,y,color='black',marker='.',linestyle='')
```

```python
plt.plot(x,y,color='black',marker='.',linestyle='')
```

```python
plt.plot(x,y,c='k',marker='.',ls='')
```

```python
plt.plot(x,y,'k.')
```

```python
plt.plot(x,y,c='b',marker='o',alpha=0.6)
```

```python
x_mean=np.mean(x)
```

```python
x_mean
```

```python
slope,intercept,r_value,p_value,std_err=stats.linregress(x-x_mean,y)
```

```python
slope,intercept
```

```python
r_value,p_value,std_err
```

```python
xx=np.linspace(800,8400,1000)
xx
```

```python
xx_fit=xx-x_mean
yy_fit=slope*xx_fit+ intercept
yy_fit
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,'k.')
#plt.plot(xx,yy,'b--',label="mean ration = %.2f" % mean_ratio)
plt.plot(xx,yy_fit,'r-',label = "linreg slope = %.2f"%slope)

plt.xlabel("No. of Contributing Users")
plt.ylabel("No. of Edits")
plt.minorticks_on()
plt.title("Forbes 200 Celberities' Wikipedia Pages")
plt.legend(loc="upper left", numpoints=1)
plt.ylim((0,27000))
plt.savefig("../fig/fig_wikipedia_users_vs_no_of_edits.png")
```
